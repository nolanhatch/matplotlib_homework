```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)

ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset

pyber_data = pd.merge(city_data, ride_data, on="city")


pyber_data.head()
# Display the data table for preview
```

```python


city_grouped = pyber_data.groupby("city")
city_grouped.head()
total_rides_city = city_grouped["ride_id"].count()
type_grouped = pyber_data.groupby("type")

#xaxis

avg_fare_city = city_grouped["fare"].mean()

dropped_df = pyber_data.drop_duplicates("city")
dropped_group = dropped_df.groupby("city")
dropped_group = dropped_group["driver_count"].sum()
driver_count = city_grouped["ride_id"].count()
color = ["yellow", "lightskyblue", "orange"]


plt.scatter( x=total_rides_city, y=avg_fare_city, s=dropped_group, alpha=0.5, color=color)

plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
plt.xticks([5,10,15,20,25,30,35,40])
plt.yticks([20,25,30,35,40])
plt.legend()
#plt.legend(handles=)
#plt.scatter(x, y, s=area, c=colors, alpha=0.5)

plt.show()
```
